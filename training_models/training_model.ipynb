```python
import cv2
from ultralytics import YOLO

cap = cv2.VideoCapture(0)
model = YOLO("../runs_nano/detect/train5/weights/best.pt")
while(cap.isOpened()):
    ret, frame = cap.read()
    if cv2.waitKey(1) & 0xFF in [ord('q'), ord('–π')] or not ret:
        break

    results = model(frame, stream=True)

    for r in results:
        frame = r.plot()

    cv2.imshow("Test tracking", frame)

cap.release()
cv2.destroyAllWindows()
```
