```python
import mediapipe as mp
import cv2
import numpy as np
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
poseDetector = mp_pose.Pose(min_detection_confidence=0.7, min_tracking_confidence=0.7)
cap = cv2.VideoCapture(0)
while(cap.isOpened()):
    ret, frame = cap.read()
    if cv2.waitKey(1) & 0xFF in [ord('q'), ord('Ð¹')] or not ret:
        break
    frame = cv2.cvtColor(np.fliplr(frame), cv2.COLOR_BGR2RGB)
    results = poseDetector.process(frame)
    if results.pose_landmarks:
        mp_drawing.draw_landmarks(
            frame,
            results.pose_landmarks,
            mp.solutions.pose.POSE_CONNECTIONS,
            mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=4, circle_radius=2),
            mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=6, circle_radius=2)
        )
    cv2.imshow('frame', cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

poseDetector.close()
cap.release()
cv2.destroyAllWindows()
```
