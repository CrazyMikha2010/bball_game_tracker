```python
import cv2
import numpy as np
```

```python
lower = np.array([2, 120, 20])
upper = np.array([7, 200, 255])
```

```python
cap = cv2.VideoCapture("videos/test1.mp4")

while True:
    ret, frame = cap.read()

    frameHSV = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(frameHSV, lower, upper)
    frame_masked = cv2.bitwise_and(frame, frame, mask=mask)
    contours, hierarchy = cv2.findContours(cv2.cvtColor(frame_masked, cv2.COLOR_BGR2GRAY), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    for contour in contours:
        area = cv2.contourArea(contour)
        x, y, w, h = cv2.boundingRect(contour)
        if area > 35 and abs(1 - (w/h)) < 0.2 :
            cv2.drawContours(frame_masked, [contour], -1, (255, 0, 0), 3)
    cv2.imshow('frame', frame_masked)
    cv2.waitKey(1)
```
